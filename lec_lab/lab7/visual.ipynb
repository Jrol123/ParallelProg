```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('output.csv')

df
```

```python
df.drop(['Velocity X', 'Velocity Y', 'Velocity Z'], axis=1, inplace=True)
df
```

```python
count_dots = 4
count_steps = 50
```

```python
coords_tuple = (0, count_dots - 1)
for i in range(count_steps):
    sdf = df.drop(["Step", "Point"], axis=1).loc[coords_tuple[0] + count_dots * i: coords_tuple[1] + count_dots * i]
    x = sdf["Position X"]
    y = sdf["Position Y"]
    z = sdf["Position Z"]
    # Creating figure
    fig = plt.figure(figsize = (10, 7))
    ax = plt.axes(projection ="3d")
    ax.set_xlim(-2, 7)  # x-axis limits
    ax.set_ylim(-4, 4)  # y-axis limits
    ax.set_zlim(-6, 6)  # z-axis limits
    
    # Creating plot
    ax.scatter3D(x, y, z, color = "blue", s=25)
    # plt.title("simple 3D scatter plot")
    # out_img = sns.scatterplot()
    scatter_fig = fig.get_figure()
 
    # use savefig function to save the plot and give
    # a desired name to the plot.
    scatter_fig.savefig(f'img/scatterplot_{i}.png')
    
    # plt.clf()
    plt.close()
```

```python
import os
from PIL import Image

# Папка, где сохранены изображения
image_folder = 'img'

# Получение списка изображений
images = []
for i in range(count_steps):
    image_path = f'{image_folder}/scatterplot_{i}.png'
    images.append(Image.open(image_path))

# Создание GIF анимации
images[0].save(f'scatter_animation_{count_steps}.gif', save_all=True, append_images=images[1:], optimize=False, duration=100, loop=0)
```
